```python
fig, axes = plt.subplots()
axes.set_xlabel('size of input')
axes.set_ylabel('time complexity')
axis.set_title('demo graph')
axes.plot(x,x)
axes.plot(x,x**2, alpha = 0.5)

axes.legend(["okay time complexity","bad time complexity"],loc = 1)
```

```python
fig,axes = plt.subplots(dpi =150)
axes.grid(True)
axes.plot(x,x+2,color = 'blue',linewidth = 0.5,linestyle='-',marker = 'o',markersize =5)
axes.plot(x,x+3,color = 'blue',linewidth = 0.5,linestyle=':',marker = '1',markerfacecolor = 'red')
axes.plot(x,x+4,color = 'blue',linewidth = 0.5,linestyle='--',marker = '+')
axes.plot(x,x+5,color = 'blue',linewidth = 0.5,linestyle='-.',marker = 's')

line, = axes.plot(x,x+7,color = "blue",linewidth = 0.5)
line.set_dashes([5,10,15,10])

line, = axes.plot(x,x,color = "black",linewidth = 0.9)
line.set_dashes([2,5,7,1])
```

```python
n = np.array([0,1,2,3,4,5])

fig,axes=plt.subplots(1,4,figsize = (16,5))
axes[0].set_title('scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x))) 

axes[1].set_title("step")
axes[1].step(n,n**2,lw =2)

axes[2].set_title("bar")
axes[2].bar(n,n**2)

axes[3].set_title("fill_between")
axes[3].fill_between(n,n**2,lw =2,color ="green")
```
