```python
import numpy as np

x = np.linspace(0,10,25)
y = x*x + 2
print(x)
print(y)

```

```python
# operator description
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5,0.1,0.8,0.8])

axis.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y, 'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize = (16,9), dpi = 100) # dots per inches

fig.add_subplot()

plt.plot(x,y,'r')
```

```python
fig, axes = plt.subplots() # use dpi
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')

axes.set_title('demo graph')

axes.plot(x,x**2)
axes.plot(x,x**3)

axes.legend(['y = x**2', 'y = x**3'],loc = 2)
```

```python
fig, axes = plt.subplots(dpi = 150) # use dpi

axes.plot(x,x**2,color = 'red',alpha = .5)
axes.plot(x,x+3, color = 'blue', alpha = .5)
axes.plot(x,x+4, color = 'green', alpha = .5)
```

```python
fig, axes = plt.subplots(dpi = 150) # use dpi

axes.plot(x,x+2,color = 'blue', linewidth = 0.5, linestyle = '-')
axes.plot(x,x+3, color = 'blue',  linewidth = 1, linestyle = '-.')
axes.plot(x,x+4, color = 'blue',  linewidth = 1.5, linestyle = ':')
axes.plot(x,x+5, color = 'blue', linewidth = 2, linestyle = '--')


line, = axes.plot(x,x+7,color = 'blue', linewidth = 0.5)
line.set_dashes([5,10,15,10])

line, = axes.plot(x,x+6, color = 'blue',  linewidth = 1)
line.set_dashes([5,10,15,20])
```

```python
fig, axes = plt.subplots(dpi = 150) # use dpi

axes.plot(x,x+2,color = 'blue', marker = 'o',markersize = 5)
axes.plot(x,x+3, color = 'blue', marker = 's')# markerfacecolor = 'red'
axes.plot(x,x+4, color = 'blue', marker = '1')
axes.plot(x,x+5, color = 'blue',marker = '+')

```

```python
fig, axes = plt.subplots(dpi = 150)
axes.plot(x,x+2,color = 'green', linewidth = 0.5, linestyle = '-')
axes.plot(x,x+3, color = 'red',  linewidth = 1, linestyle = '-.')
axes.plot(x,x+4, color = 'blue',  linewidth = 1.5, linestyle = ':')
axes.grid(True)
```

# 2D Graphic in matplotlib

```python
n = np.array([0,1,2,3,4,5])

fig,axes = plt.subplots(1,4,figsize = (16,5))
axes[0].set_title('scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw = 2)

axes[2].set_title("bar")
axes[2].bar(n,n**2, align = 'center')

axes[3].set_title('fill_between')
axes[3].fill_between(x,x**2,color = 'green')
```

# histogram 

```python
n = np.random.randn(100000)
fig,axes = plt.subplots(1,2, figsize = (12,4),dpi = 300)
axes[0].set_title('defualt histogram')
axes[0].hist(n)

axes[1].set_title('cumulative detailed histogram')
axes[1].hist(n,cumulative = True, bins = 10)
```

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
x,y = np.meshgrid(x,y)

z1 = np.exp(-x**2 - y**2)
z2 = np.exp(-(x-1)**2 - (y-1)**2)
z = (z1 - z2) * 2

print(x)
print(y)
print(z)
```

```python
fig, axes = plt.subplots()
cs = axes.contour(x,y,z)
axes.clabel(cs,inline = 1, fontsize = 10)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (14,6))

axes = fig.add_subplot(1,2,1,projection = '3d')
axes.plot_surface(x,y,z,rstride = 4, cstride = 4, linewidth = 0)
```
